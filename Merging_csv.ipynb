# Setup

```python
import pandas as pd
```

# Read in CSVs

```python
alcohol_csv = "HappinessAlcoholConsumption.csv"
country_abbr_csv = "country_abbr.csv"
countries_csv = 'countries.csv'

coords_df = pd.read_csv(countries_csv)
alcohol_df = pd.read_csv(alcohol_csv)
country_df = pd.read_csv(country_abbr_csv)
```

# Rename 'Name' Column to 'Country'

```python
country_df.columns = ["Country", "Abbreviation"]

alcohol_df
# country_df
```

```python
coords_df.head()
coords_df.dropna()
coords_df = coords_df.drop(["country"], axis = 1)
coords_df = coords_df.rename(columns = {'name': 'Country'})
coords_df.head()
```

```python
country2_df = pd.merge(country_df, coords_df, on = "Country")
country2_df.head()
```

```python
final_alcohol_df = pd.merge(alcohol_df,
                            country2_df,
                            on="Country")
final_alcohol_df
```

# Export Final DataFrame as CSV

```python
final_alcohol_df.to_csv("FinalAlcohol.csv")
```
