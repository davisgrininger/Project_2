# Setup

```python
import pandas as pd
```

# Read in CSVs

```python
alcohol_csv = "HappinessAlcoholConsumption.csv"
country_abbr_csv = "country_abbr.csv"

alcohol_df = pd.read_csv(alcohol_csv)
country_df = pd.read_csv(country_abbr_csv)
```

# Rename 'Name' Column to 'Country'

```python
country_df.columns = ["Country", "Abbreviation"]

alcohol_df
# country_df
```

```python
final_alcohol_df = pd.merge(alcohol_df,
                            country_df,
                            on="Country")
final_alcohol_df
```

# Export Final DataFrame as CSV

```python
final_alcohol_df.to_csv("FinalAlcohol.csv")
```
